#Load Models

```python
model_type = "glioma"
#model_type = "meningioma"
#model_type = "no_tumor"
#model_type = "pituitary_new"
```

```python
from diffusers import DDPMPipeline

pipeline = DDPMPipeline.from_pretrained(f"emendes3/cancer_diffusion_model_{model_type}")
```

```python
import torch
import os
def generate_images(epoch, pipeline, model_type):
    images = pipeline(
        batch_size = 16, 
        generator=torch.manual_seed(epoch),
    ).images

    # Save the images
    test_dir = model_type
    os.makedirs(test_dir, exist_ok=True)
    for i,img in enumerate(images):
      img.save(f"{test_dir}/{epoch:04d}_{i}.png")
```

```python
for i in range (1, 2):
  generate_images(i, pipeline, model_type)
```
